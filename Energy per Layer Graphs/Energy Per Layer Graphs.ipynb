```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ultdataset.csv")
```

```python
df_high_energy = df[df["Beam Energy"]>= 200]
```

```python
df_high_energy.reset_index(inplace = True)
```

```python
df_high_energy
```

```python
layers = df_high_energy.drop(["Beam Energy", "Particle", "index"], axis=1, inplace = False)
```

```python
layers
```

```python
for i in range(300):
    plt.bar(x = np.array(range(1, 41)), height = np.array(layers.loc[i]))
    plt.xlabel("Layers")
    plt.ylabel("Energy")
    plt.title("Energy per Layer (MIP) for {particle} at index {index}".format(
        particle = df_high_energy.loc[i]["Particle"], index =i))
    plt.show()
```
